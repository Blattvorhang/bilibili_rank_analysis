```python
import nest_asyncio
from bilibili_api import rank, sync
import pandas as pd

# 使得jupyter notebook支持asyncio
nest_asyncio.apply()
```

```python
rank_info = sync(rank.get_rank())
rank_info
```

```python
# 读取本地数据，保证分析的一致性
import rank_data
rank_info = rank_data.rank_dict
rank_info.keys()
```

```python
rank_info['list'][0].keys()
```

```python
rank_info['list'][0]
```

```python
title_list = []
pubdate_list = []
duration_list = []
owner_list = []
owner_mid_list = []
view_list = []
danmaku_list = []
reply_list = []
favorite_list = []
coin_list = []
share_list = []
his_rank_list = []
like_list = []
dimension_list = []
pub_location_list = []
bvid_list = []

for video in rank_info['list']:
    title_list.append(video['title'])
    pubdate_list.append(video['pubdate'])
    duration_list.append(video['duration'])
    owner_list.append(video['owner']['name'])
    owner_mid_list.append(video['owner']['mid'])
    view_list.append(video['stat']['view'])
    danmaku_list.append(video['stat']['danmaku'])
    reply_list.append(video['stat']['reply'])
    favorite_list.append(video['stat']['favorite'])
    coin_list.append(video['stat']['coin'])
    share_list.append(video['stat']['share'])
    his_rank_list.append(video['stat']['his_rank'])
    like_list.append(video['stat']['like'])
    dimension_list.append(str(video['dimension']['width']) + 'x' +
                          str(video['dimension']['height']))
    pub_location_list.append(video['pub_location'])
    bvid_list.append(video['bvid'])
```

```python
df = pd.DataFrame({
    '视频标题': title_list,
    '发布时间': pubdate_list,
    '时长': duration_list,
    'up主': owner_list,
    'up主mid': owner_mid_list,  # 'mid'是b站用户的唯一标识符，可以用来获取用户信息
    '播放量': view_list,
    '弹幕数': danmaku_list,
    '评论数': reply_list,
    '收藏量': favorite_list,
    '投币量': coin_list,
    '转发数': share_list,
    '历史排名': his_rank_list,
    '点赞数': like_list,
    '分辨率': dimension_list,
    '发布地点': pub_location_list,
    'BV号': bvid_list
})
```
