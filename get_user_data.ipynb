```python
sync(u.get_relation_info())
```

```python
mid_list = df['up主mid'].map(int).tolist()
name_list = df['up主'].tolist()

user_df = pd.DataFrame({'UID': mid_list, '昵称': name_list})
user_df.drop_duplicates(subset='UID', inplace=True)  # 保证UID是唯一的
user_df.set_index('UID', inplace=True)
user_df.head()
```

```python
len(user_df)
```

```python
info_list = []

# 异步获取所有up主的信息
for mid, name in zip(mid_list, name_list):
    u = user.User(mid)
    info = await u.get_user_info()
    info_list.append(info)
    print(f'“{name}”的信息已获取')
```

```python
relation_list = []

# 异步获取所有up主的关注和粉丝信息
for mid, name in zip(mid_list, name_list):
    u = user.User(mid)
    relation = await u.get_relation_info()
    relation_list.append(relation)
    print(f'“{name}”的关注和粉丝信息已获取')
```

```python
len(info_list), len(relation_list)
```

```python
for info in info_list:
    mid = info['mid']
    user_df.loc[mid, '性别'] = info['sex']
    user_df.loc[mid, '签名'] = info['sign']
    user_df.loc[mid, '等级'] = info['level']
    user_df.loc[mid, '头衔'] = info['official']['title']
    
for relation in relation_list:
    mid = relation['mid']
    user_df.loc[mid, '关注数'] = relation['following']
    user_df.loc[mid, '粉丝数'] = relation['follower']
    
user_df.head()
```

```python
user_df.to_csv('./data/bilibili_user.csv', encoding='utf-8-sig')
```
